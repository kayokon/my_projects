```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from scipy.integrate import odeint
```

```python
L1 = input('Set the length of string 1 (m): ')
L1 = float(L1) 
L2 = input('Set the length of string 2 (m): ')
L2 = float(L2) 
m1 = input('Set the value of mass 1 (kg): ') 
m1 = float(m1)
m2 = input('Set the value of mass 1 (kg): ') 
m2 = float(m2)
int_phi1 = input('input initial angle (degrees) of mass 1: ')
int_phi1 = np.radians(float(int_phi1))
int_phi2 = input('input initial angle (degrees) of mass 2: ')
int_phi2 = np.radians(float(int_phi2))
g = 9.8

initial_angle1 = [int_phi1, int_phi2, 0.0, 0.0]
initial_angle2 = [int_phi1+0.00001, int_phi2, 0.0, 0.0]
t = np.arange(0.0, 50, 0.05)

angle = np.zeros(4)
phi1 = angle[0]
phi2 = angle[1]
phi1_dot = angle[2]
phi2_dot = angle[3]

def ODE(angle,t):

    phi1_dot = angle[2]
    phi2_dot = angle[3]
    
    delta1 = angle[0] - angle[1]
    M = m1 + m2
    phi1_dot_dot = (-m2*L2*angle[3]**2*np.sin(delta1)
               - M*g*np.sin(angle[0])
               - m2*L1*angle[2]**2*np.cos(delta1)*np.sin(delta1)
               + m2*g*np.sin(angle[1])*np.cos(delta1)) / (M*L1 - m2*L1*np.cos(delta1)**2)

    phi2_dot_dot = (M*L1*angle[2]**2*np.sin(delta1)
               - M*g*np.sin(angle[1])
               + m2*L2*angle[3]**2*np.cos(delta1)*np.sin(delta1) 
               + M*g*np.cos(delta1)*np.sin(angle[0])) / (M*L2 - m2*L2*np.cos(delta1)**2)

    return phi1_dot, phi2_dot, phi1_dot_dot, phi2_dot_dot


sol_1 = odeint(ODE, initial_angle1, t)
sol_2 = odeint(ODE, initial_angle2, t)

x1_1 = L1*np.sin(sol_1[:,0])
y1_1 = -L1*np.cos(sol_1[:,0])
x2_1 = L2*np.sin(sol_1[:,1]) + x1_1
y2_1 = -L2*np.cos(sol_1[:,1]) + y1_1

x1_2 = L1*np.sin(sol_2[:,0])
y1_2 = -L1*np.cos(sol_2[:,0])
x2_2 = L2*np.sin(sol_2[:,1]) + x1_2
y2_2 = -L2*np.cos(sol_2[:,1]) + y1_2

fig = plt.figure()
ax = fig.add_subplot(111,xlim = [-(L1 + L2 +2),(L1 + L2 +2)], 
                     ylim = [-(L1 + L2 + 2),(L1 + L2 +2)])

ax.patch.set_facecolor('black')

string_1, = ax.plot([], [],c="blue",lw=1)
mass1_1,= ax.plot([], [], 'o-', markersize=0.0+m1,c="blue")
mass2_1, = ax.plot([], [], 'o-', markersize=0.0+m2,c="blue")
trace_1, = ax.plot([], [], '.', markersize=1,c="aqua")

string_2, = ax.plot([], [], c="pink",lw=1)
mass1_2,= ax.plot([], [], 'o-', markersize=0+m1,c="pink")
mass2_2, = ax.plot([], [], 'o-', markersize=0+m2,c="pink")
trace_2, = ax.plot([], [], '.', markersize=1,c="lightpink")

time_template = 'time = %.1fs'
time_text = ax.text(0.05, 0.9, '', transform=ax.transAxes, color='red')

def draw(i):
    string_x1 = [0, x1_1[i], x2_1[i]]
    string_y1 = [0, y1_1[i], y2_1[i]]
    mass1_x1 = [x1_1[i]]
    mass2_x1 = [x2_1[i]]
    mass1_y1 = [y1_1[i]]
    mass2_y1 = [y2_1[i]]
    trace_x1 = [x2_1[:i]]
    trace_y1 = [y2_1[:i]]  
    
    string_x2 = [0, x1_2[i], x2_2[i]]
    string_y2 = [0, y1_2[i], y2_2[i]]
    mass1_x2 = [x1_2[i]]
    mass2_x2 = [x2_2[i]]
    mass1_y2 = [y1_2[i]]
    mass2_y2 = [y2_2[i]]
    trace_x2 = [x2_2[:i]]
    trace_y2 = [y2_2[:i]]
    
    string_1.set_data(string_x1, string_y1)
    mass1_1.set_data(mass1_x1, mass1_y1)
    mass2_1.set_data(mass2_x1, mass2_y1)
    trace_1.set_data(trace_x1, trace_y1)
    string_2.set_data(string_x2, string_y2)
    mass1_2.set_data(mass1_x2, mass1_y2)
    mass2_2.set_data(mass2_x2, mass2_y2)
    trace_2.set_data(trace_x2, trace_y2)
    
    time_text.set_text(time_template % (i*0.05))
    
    return  string_1, mass1_1, mass2_1, trace_1, string_2, mass1_2, mass2_2, trace_2, time_text

%matplotlib inline
from IPython.display import HTML
get_draw = animation.FuncAnimation(fig, draw, np.arange(0, len(sol_1)), interval=30, blit=True)
HTML(get_draw.to_html5_video())
#get_draw = animation.FuncAnimation(fig, draw, np.arange(0, len(sol_1)), interval=30, blit=True)
plt.show()
```
